```python
ar1= pd.DataFrame(air['COORDINATES'].tolist(), columns={'lat', 'lon'})
ar1.insert(2, 'CO', air['CO'])
ar1.insert(3, 'O3', air['O3'])
if1= pd.DataFrame(inf['COORDINATES'].tolist(), columns={'lat', 'lon'})
if1.insert(2, 'inf', inf['infections'])
new = pd.concat([if1, ar1], ignore_index=True, sort=False).sort_values(['lat', 'lon']).reset_index()
new = new.drop_duplicates(subset=['lat', 'lon'])
new = new.drop(columns={"index"})

new['CO'] = new['CO'].interpolate(method='nearest')
new['O3'] = new['O3'].interpolate(method='nearest')
new['inf'] = new['inf'].fillna(0)
new['CO'] = (new['CO']-1)/(4-1)
new['O3'] = (new['O3']-0.070118)/(13.359422-0.070118)
new['CH4'] = met['CH4']/1.919921  
new['Month'] = 1
new = new.loc[(new['inf'] > 0)]
```
