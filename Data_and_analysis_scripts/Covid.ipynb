```python
covid = covid.loc[(covid['latitude'] != 'NaN') & (covid['longitude'] != 'NaN')]
```

```python
cod = covid['date_confirmation'].str.replace('.', '')
```

```python
cod = cod.str[2:-4]
```

```python
covid.insert(3, "date", cod)
```

```python
covid
```

```python
covid = covid.drop(columns = 'date_confirmation')
```

```python
covid
```

```python
covid['latitude'] = pd.to_numeric(covid['latitude'])
```

```python
covid['longitude'] = pd.to_numeric(covid['longitude'])
```

```python
covid
```

```python
covid = covid.loc[(covid["latitude"] >= 15) &(covid["latitude"] <= 60) &(covid["longitude"] >=70)& (covid["longitude"] <= 135)]
```

```python
covid = covid.rename(columns={'latitude':'lat', 'longitude': 'lon'})
```

```python
covid1 = covid.loc[(covid['date'] == '01')]
covid2 = covid.loc[(covid['date'] == '02')]
covid3 = covid.loc[(covid['date'] == '03')]
covid4 = covid.loc[(covid['date'] == '04')]
covid5 = covid.loc[(covid['date'] == '05')]
covid6 = covid.loc[(covid['date'] == '06')]
```

```python
covid1.drop('date', axis=1)
covid2.drop('date', axis=1)
covid3.drop('date', axis=1)
covid4.drop('date', axis=1)
covid5.drop('date', axis=1)
covid6.drop('date', axis=1)
```

```python
index1= ['lat', 'lon']
```

```python
covid1= covid1.pivot_table(index=index1, aggfunc='size')
covid2= covid2.pivot_table(index=index1, aggfunc='size')
covid3= covid3.pivot_table(index=index1, aggfunc='size')
covid4= covid4.pivot_table(index=index1, aggfunc='size')
covid5= covid5.pivot_table(index=index1, aggfunc='size')
covid6= covid6.pivot_table(index=index1, aggfunc='size')
```

```python
covid1= covid1.reset_index()
covid2= covid2.reset_index()
covid3= covid3.reset_index()
covid4= covid4.reset_index()
covid5= covid5.reset_index()
covid6= covid6.reset_index()
```

```python
covid1= covid1.rename(columns={0:"infections"})
covid2= covid2.rename(columns={0:"infections"})
covid3= covid3.rename(columns={0:"infections"})
covid4= covid4.rename(columns={0:"infections"})
covid5= covid5.rename(columns={0:"infections"})
covid6= covid6.rename(columns={0:"infections"})
```

```python
new = pd.concat([covid1, covid2, covid3, covid4, covid5, covid6], ignore_index=True, sort=False).sort_values(['lat', 'lon']).reset_index()
```
