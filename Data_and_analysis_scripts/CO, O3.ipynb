```python
car = dataset['CO'].to_dataframe()
oz = dataset['O3'].to_dataframe()
car = car * 1000000000
oz = oz * 1000000
oz=oz.reset_index()
car=car.reset_index()
car = car.loc[(car["lat"] >= 15) &(car["lat"] <= 60) &(car["lon"] >=70)& (car["lon"] <= 135)]
oz = oz.loc[(oz["lat"] >= 15) &(oz["lat"] <= 60) &(oz["lon"] >=70)& (oz["lon"] <= 135)]
car['O3'] = oz['O3']
car = car.drop(['lev', 'time'], axis =1)
car = car.drop_duplicates(subset={'lat', 'lon'}, keep='first')
```
