```python
import pandas as pd
import numpy as np
import xarray as xr
import json

ds = xr.open_dataset("MERRA2_400.tavg1_2d_chm_Nx.20200601.nc4") # take COCL and TO3
df = ds.to_dataframe()
print(df)
del df['COEM']
del df['COLS']
del df['COPD']
del df['COCL']
del df['LWI']
# china is between 15-60* N, 70-135 E

df = df.reset_index()
del df['time']
df = df.loc[(df["lat"] >= 15) &(df["lat"] <= 60) &(df["lon"] >=70)& (df["lon"] <= 135)]
df = df.drop_duplicates(subset=['lat','lon'])
df = df[['lat', 'lon', 'COSC', 'TO3']]
data = []
for row in df.itertuples(index = False):
    point_data = {
        'COORDINATES': [row[1], row[0]],
        'CO': row[2],
        'O3': row[3],
    }
    data.append(point_data)

with open('data_CO_O3_2020_06.json', 'w') as f:
    json.dump(data, f,indent=2)
```

```python
import pandas as pd
import numpy as np
import xarray as xr
from pyhdf.SD import SD, SDC
import pprint
import json

file_name = 'AIRS.2020.05.31.L3.RetStd_IR001.v7.0.3.0.G20154092215.hdf'
readTO = 'data_CH4_2020_06.json'
file = SD(file_name,SDC.READ)
print(file.info())
datasets_dic = file.datasets()

#for idx,sds in enumerate(datasets_dic.keys()):
    #print(idx,sds)

sds_obj = file.select('CH4_VMR_A')
data = sds_obj.get()
#print(sds_obj.attributes())

pprint.pprint( sds_obj.attributes() )

print(data.shape) 

#pn = data.to_numpy()

df = pd.DataFrame(data[0])
df = df.replace(to_replace={-9999:0},value=None)
dt180 = {}
for i in range(0,180):
    dt180[i]=i-90
df = df.rename(index=dt180)
dt360= {}
for i in range(0,360):
    dt360[i]=i-180
df = df.rename(columns=dt360)


df = df.stack()
df = df.reset_index()
df = df.rename(columns = {'level_0':'lat','level_1':'lon',0:'CH4'})
df['CH4'] = df['CH4'].apply(lambda x: x*1000000)
                    
df = df.loc[(df["lat"] >= 15) &(df["lat"] <= 60) &(df["lon"] >=70)& (df["lon"] <= 135)]

k=[]
for i in range(140,271):
    k.append(i/2)

nans = []
for j in range(15,61):
    for i in range(70*8,135*8):
        if i/8 != round(i/8):
            nans.append([j,i/8,np.nan])
nans = pd.DataFrame(data=nans)
nans = nans.rename(columns={0:'lat',1:'lon'})
dfexp1 = pd.concat([df, nans], ignore_index=True, sort=False).sort_values(['lat', 'lon']).reset_index()
dfexp1 = dfexp1.drop(columns=['index',2])
dfexp1 = dfexp1.set_index(['lon','lat'],drop=True)


dfexp1 = dfexp1.interpolate()
dfexp1 = dfexp1.reset_index()




dfexp1 = dfexp1.set_index('lat')
nans = []
for j in range(70*8,135*8):
    for i in range(15,60):
        nans.append([float(i)+0.5,j/8,np.nan])
nans = pd.DataFrame(data=nans)
nans = nans.rename(columns={0:'lat',1:'lon'})
dfexp1 = dfexp1.reset_index()
dfexp1 = pd.concat([dfexp1, nans], ignore_index=True, sort=False).sort_values(['lat', 'lon']).reset_index()
dfexp1 = dfexp1.drop(columns=['index',2])
dfexp1 = dfexp1.set_index(['lon','lat'],drop=True)
dfexp1 = dfexp1.sort_index(axis=0,level=0)
dfexp1 = dfexp1.interpolate()
dfexp1 = dfexp1.reset_index()
dfexp1 = dfexp1.set_index(['lat','lon'],drop=True)
dfexp1 = dfexp1.reset_index()

dfexp1 = dfexp1[['lat', 'lon', 'CH4']]
dfexp1 = dfexp1.loc[(dfexp1['lon'] % .625 == 0)]
dfexp1 = dfexp1.set_index(['lat','lon'])
dfexp1 = dfexp1.sort_index(axis=0,level=0)
dfexp1 = dfexp1.reset_index()


data = []
for row in dfexp1.itertuples(index = False):
    point_data = {
        'COORDINATES': [row[0], row[1]],
        'CH4': row[2],
    }
    data.append(point_data)

with open(readTO, 'w') as f:
    json.dump(data, f,indent=2)
```
